```python
from pandas.plotting import scatter_matrix
scatter_matrix(iris, alpha=0.2, figsize=(8, 8), diagonal='kde')
```

```python
from pltnine import *
```

```python
p = ggplot(aes(x='SepalLengthCm', y='SepalWidthCm'), data=iris) + geom_jitter()
p
```

```python
p = ggplot(aes(x='SepalLengthCm', y='SepalWidthCm', 
           color='Species', size='PetalLengthCm'), 
           data=iris) + geom_jitter()
p
```

```python
p = ggplot(aes(x='SepalLengthCm', color='Species'), 
           data=iris) + geom_density()
p
```

```python
p = ggplot(aes(x='SepalLengthCm', y='SepalWidthCm'), 
           data=iris) + geom_point(alpha=0.8, color='steelblue') + \
           facet_wrap('Species')
p
```

### Regresión lineal

Uno de los algoritmos más básicos y del cual se derivan otros es la regresión lineal. El método más común para resolverlo es por gradiente descendiente. El cual se expresa a continuación:

$\hat{y} = \theta x$ <br>
$\theta_{t+1} = \theta_{t} - \alpha  \sum{(\hat{y}(x_{i}) - y(x_{i}))} x_{i}$

Donde $x$ y $y$ son variables conocidas y suponemos que y depende de $x$, $\theta$ son los parámetros del modelo, $\hat{y}$ es la salida estimada y $\alpha$ es un factor de actualización.

La función de actualización es consecuencia de la función de costo de mínimos cuadrados. <br>

$\hat{\theta}=argmin([\theta X - Y]^{2})$ <br>
$\hat{\theta}=\nabla(\theta X - Y)^{2}$
